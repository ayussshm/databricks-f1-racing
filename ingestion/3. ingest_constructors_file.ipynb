#### Step 1- Read the JSON file using Spark

```python
constructors_schema = "constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRING"

constructors_df = spark.read.schema(constructors_schema)\
                            .json(f'{raw_folder_path}/constructors.json')
```

#### Step 2: Drop the url columm

```python
constructors_dropped_df = constructors_df.drop("url")
```

#### Step 3: Rename columns and add ingestion date

```python
from pyspark.sql.functions import lit

constructors_final_df = add_ingestion_date(constructors_dropped_df.withColumnRenamed("constructorId", "constructor_id")\
                                              .withColumnRenamed("constructorRef", "constructor_ref")\
                                              .withColumn('data_source', lit(v_data_source)))
```

#### Step 4: Write data to parquet file

```python
constructors_final_df.write\
                    .mode("overwrite")\
                    .parquet(f'{processed_folder_path}/constructors')
                
```
