# Access Azure Data Lake using access keys
1. Set spark config fs.azure.account.key
2. List files from demo container
3. Read data from circuits.csv file

We need to first obtain secrets using secret scopes

```python
formula1dl_account_key = dbutils.secrets.get(scope='formula1-scope', key='formula1dl-account-key')
```

```python
storage_account = "formula1dl244"
```

```python
spark.conf.set(f"fs.azure.account.key.{storage_account}.dfs.core.windows.net",
               formula1dl_account_key)
```

```python
container = "demo"
display(dbutils.fs.ls(f"abfss://{container}@{storage_account}.dfs.core.windows.net"))
```

```python
display(spark.read.csv(f"abfss://demo@formula1dl244.dfs.core.windows.net/circuits.csv"))
```
