### Learning Objectives

2. Create External table using SQL

```sql
%python
race_results_df.write.format('parquet').option('path', f'{presentation_folder_path}/race_results_ext_py').saveAsTable('demo.race_results_ext_py')
```

```sql
DESC EXTENDED demo.race_results_ext_py
```

```sql
CREATE TABLE demo.race_results_ext_sql
(
race_year INT,
race_name STRING,
race_date TIMESTAMP,
circuit_location STRING,
driver_name STRING,
driver_number INT,
driver_nationality STRING,
team STRING,
grid INT,
fastest_lap INT,
race_time STRING,
points INT,
position INT,
ingestion_date TIMESTAMP
)
USING PARQUET
LOCATION '/mnt/formula1dl244/presentation/race_results_ext_sql'
```

```sql
SHOW TABLES in demo;
```

```sql
INSERT INTO TABLE demo.race_results_ext_sql
SELECT * 
  FROM demo.race_results_ext_py WHERE race_year = 2020;
```

```sql
SELECT COUNT(*) FROM demo.race_results_ext_sql
```

```sql
SHOW tables in demo;
```

```sql
DROP TABLE demo.race_results_ext_sql
```

```sql
SHOW tables in demo;
```

### Views on tables
1. Create Temp View
2. Create Global Temp View
3. Create Permanent View

```sql
CREATE OR REPLACE TEMP VIEW v_race_results AS
SELECT *
FROM 
demo.race_results_python
WHERE race_year = 2010;
```

```sql
SELECT * FROM v_race_results;
```

```sql
CREATE OR REPLACE GLOBAL TEMP VIEW gv_race_results AS
SELECT *
FROM demo.race_results_python
WHERE race_year = 2010;
```

```sql
SELECT * FROM global_temp.gv_race_results
```

```sql
SHOW TABLES in global_temp;
```

```sql
CREATE OR REPLACE VIEW demo.pv_race_results AS
SELECT *
FROM demo.race_results_python
WHERE race_year = 2000;
```

```sql
SHOW TABLES;
```
