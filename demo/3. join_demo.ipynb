```python
circuits_df = spark.read.parquet(f"{processed_folder_path}/circuits")
circuits_df = circuits_df.filter(circuits_df.circuit_id < 70)
```

```python
races_df = spark.read.parquet(f"{processed_folder_path}/races")
races_df = races_df.filter(races_df.race_year == '2019')
```

```python
display(circuits_df)
```

```python
display(races_df)
```

### Inner Join

```python
race_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "inner").select(circuits_df.name.alias('circuit_name'), circuits_df.location, circuits_df.country, races_df.name.alias('race_name'), races_df.round)
```

```python
display(race_circuits_df)
```

### Left Outer Join

```python
race_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "left").select(circuits_df.name.alias('circuit_name'), circuits_df.location, circuits_df.country, races_df.name.alias('race_name'), races_df.round)
```

```python
display(race_circuits_df)
```

### Right Outer Join

```python
race_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "right").select(circuits_df.name.alias('circuit_name'), circuits_df.location, circuits_df.country, races_df.name.alias('race_name'), races_df.round)
```

```python
display(race_circuits_df)
```

### Full Outer Join

```python
race_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "outer").select(circuits_df.name.alias('circuit_name'), circuits_df.location, circuits_df.country, races_df.name.alias('race_name'), races_df.round)
```

```python
display(race_circuits_df)
```

### Semi Join

```python
race_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "semi")
```

```python
display(race_circuits_df)
```

### Anti Joins

```python
race_circuits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "anti")
```

```python
display(race_circuits_df)
```

### Cros Joins

```python
race_circuits_df = races_df.crossJoin(circuits_df)
```

```python
display(race_circuits_df)
```
